```python
import pandas as pd
import json
from glob import glob
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
data_files = glob('json_data_files/*.json')
data_files
```

```python
with open(data_files[0], 'r') as myfile:
    data=myfile.read()
    
obj = json.loads(data)
print(len(obj[0]))
dict_of_cars = obj[0]
id_list = list(dict_of_cars.keys()) 

dict_of_cars[id_list[0]]
```

```python
from datetime import datetime,timedelta
```

```python
car_df = pd.DataFrame.from_dict(dict_of_cars, orient='index').reset_index()
car_df['age'] = datetime.now().year - car_df.year
```

```python
make_counts = car_df.make.value_counts()
make_counts[make_counts>10].plot(kind = 'bar',figsize=(20,10),title = "CAR COUNTS")
plt.figure()
car_df.groupby(['make']).mean().price.sort_values().plot(kind = 'barh',figsize=(20,15) ,title = "AVERAGE PRICE")
plt.figure()
car_df.groupby(['make']).mean().mileage.sort_values().plot(kind = 'barh',figsize=(20,15) ,title = "AVERAGE KMs")
```

```python
import plotly.express as px
```

```python
beemer = car_df.loc[car_df.make == 'BMW']
def get_class(val):
        return '{} series'.format(val[0])
beemer['series'] = beemer.model.apply(get_class)
beemer['avg_kms_per_year'] = round(beemer.mileage/beemer.age,2)
beemer.head()
```

```python
beemer.select_dtypes(exclude='object')
```

```python
import numpy as np
```

```python
beemer = beemer.replace([np.inf, -np.inf], np.nan)
beemer = beemer.dropna()
px.scatter(beemer,x= 'avg_kms_per_year' , y = 'price',color= 'series',hover_name = 'deriviative',trendline = 'ols',facet_row='series',height = 2000)
```
