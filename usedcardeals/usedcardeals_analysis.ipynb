```python
from lxml import html
import json
import pprint
import re
```

```python
with open('usedcardealer.json', 'r') as myfile:
    data=myfile.read()
    
obj = json.loads(data)
pages_html = obj[0]['Response']
```

```python
tree = html.fromstring(pages_html)

"data-detailurl="/view/used-vehicle/alfa-romeo/giulietta/giulietta-14tbi-distinctive-auto/1885612/" '
             'id="vid1885612" data-vehicleid="1885612" data-type="Used" '
             'data-make="Alfa Romeo" data-model="Giulietta 1.4TBi Distinctive '
             'auto" data-year="2012" data-mileage="129,000" data-price="R '
             '129,999"
```

```python
 tree.xpath('//div[@class="page-header"]//small/text()')[0].split()[0]
```

```python
ID = tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-vehicleid')
make = tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-make')
def get_model(val):
    return val.split()[0]
model_dir = tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-model')
model = list(map(get_model,model_dir ))
def get_deriv(val):
    return " ".join(val.split()[1:])
deriviative = list(map(get_deriv,model_dir ))
capacity = [re.findall("\d+\.\d+|\d+\.\d+\D+", phrase) for phrase in deriviative]
year = tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-year')
mileage = tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-mileage')
price  =tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-price')
url = tree.xpath('//div[contains(@class,"vehicle-list-block")]/@data-detailurl')
```

```python
from datetime import datetime
usedcardeals = {index : { 'year':y , 'make':mk , 'model': m , 'deriviative':d,'mileage':ml  , 'price': p ,'engine': c , 'url':u ,'date':datetime.strftime(datetime.today(),'%Y-%m-%d') } 
           for index,y,mk,m,d,ml,p,c,u in zip(ID,year,make,model,deriviative,mileage,price,capacity,url) }
```
